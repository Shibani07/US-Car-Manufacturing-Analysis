```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('car_manufacturing.csv')
df
```

```python
del df['Unnamed: 0']
del df['vin']
del df['lot']
del df['condition']
```

```python
df.head()
```

# **Top 5 Car brands in terms of number of cars**

```python
df['brand'].unique()
```

```python
data = []
#number of cars of all the brands
for brand in df['brand'].unique():
    data.append([brand , len(df[df['brand']  == brand])])

data
```

```python
data = pd.DataFrame(data,columns = ['brand','freq'])
data
```

```python
#for top 5 car brands
data = data.sort_values(by='freq', ascending=False).head()
data
```

```python
data.plot(x='brand', y='freq', kind='bar')
```

# **Car Industry Trend**

```python
data1 = []

for year in df['year'].unique():
  data1.append([year, len(df[df['year']==year])])

data1 = pd.DataFrame(data1, columns=['year','freq'])
data1
```

```python
data1 = data1.sort_values(by='year', ascending = True)
data1
```

```python
data1.plot(x = 'year', y = 'freq', kind = 'line')
```

# **Most Populat Car Colours**

```python
data2 = []

for color in df['color'].unique():
    data2.append([color , len(df[df['color']  == color])])

data2 = pd.DataFrame(data2,columns = ['color','freq'])
```

```python
data2 = data2.sort_values(by='freq', ascending=False)
data2
```

```python
data2.plot(x = 'color', y = 'freq', kind = 'bar', figsize = (20,5))
```

# **Which car brand is having Maximum Average Price**

```python
data3 = df.groupby('brand').mean()['price']
data3
```

```python
data3 = data3.sort_values(ascending = False)
data3
```

```python
data3.plot(x='brand', kind='bar')
```

# **Which State is having most expensive cars**

```python
data4 = df.groupby('state').mean()['price'].sort_values(ascending = False)
data4
```

```python
data4.plot.bar(figsize = (15,5))
```
